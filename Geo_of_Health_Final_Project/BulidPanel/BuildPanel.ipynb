### איחוד של הכנסה

```python
income_selected_years_to2014 = [2013, 2014]
income_selected_years_to2016 = [2015, 2016]
income_selected_years_to2018 = [2017, 2018]
income_selected_years_to2020 = [2019, 2020]
```

```python
# הסרת עמודות כפולות לפני המיזוג
columns_to_drop_income = ['ORG_NAME', 'tot_patients_per_PRACTICE_CODE']
merged_after_Deprivation_df = merged_after_Deprivation_df.drop(columns=columns_to_drop_income, errors='ignore')

merged_incometo2014_df = pd.merge(merged_after_Deprivation_df[merged_after_Deprivation_df['YEAR'].isin(income_selected_years_to2014)], Income_2014_df, on='PRACTICE_CODE', how='left')
merged_incometo2016_df = pd.merge(merged_after_Deprivation_df[merged_after_Deprivation_df['YEAR'].isin(income_selected_years_to2016)], Income_2016_df, on='PRACTICE_CODE', how='left')
merged_incometo2018_df = pd.merge(merged_after_Deprivation_df[merged_after_Deprivation_df['YEAR'].isin(income_selected_years_to2018)], Income_2018_df, on='PRACTICE_CODE', how='left')
merged_incometo2020_df = pd.merge(merged_after_Deprivation_df[merged_after_Deprivation_df['YEAR'].isin(income_selected_years_to2020)], Income_2020_df, on='PRACTICE_CODE', how='left')

merged_after_income_df = pd.concat([merged_incometo2014_df, merged_incometo2016_df, merged_incometo2018_df, merged_incometo2020_df], ignore_index=True)

```

### איחוד של המגדר וגיל

```python
# מיזוג כל השנים הרלוונטיות עם הקובץ Gender
merged_Gender2019_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2019], Gender_2019_df, on='PRACTICE_CODE', how='left')
merged_Gender2018_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2018], Gender_2018_df, on='PRACTICE_CODE', how='left')
merged_Gender2017_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2017], Gender_2017_df, on='PRACTICE_CODE', how='left')
merged_Gender2016_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2016], Gender_2016_df, on='PRACTICE_CODE', how='left')
merged_Gender2015_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2015], Gender_2015_df, on='PRACTICE_CODE', how='left')
merged_Gender2014_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2014], Gender_2014_df, on='PRACTICE_CODE', how='left')
merged_Gender2013_df = pd.merge(merged_after_income_df[merged_after_income_df['YEAR'] == 2013], Gender_2013_df, on='PRACTICE_CODE', how='left')

merged_Gender2019_df.info()
#merged_Gender2018_df.info()
#merged_Gender2017_df.info()
#merged_Gender2016_df.info()
#merged_Gender2015_df.info()
#merged_Gender2014_df.info()
#merged_Gender2013_df.info()



merged_all_df = pd.concat([merged_Gender2019_df, merged_Gender2018_df, merged_Gender2017_df, merged_Gender2016_df, merged_Gender2015_df, merged_Gender2014_df, merged_Gender2013_df], ignore_index=True)
```
